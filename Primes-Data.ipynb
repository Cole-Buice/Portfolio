```python
number_data = pd.DataFrame()
number_data['number'] = range(1, 500001)
```

```python
%%time
number_data[['x', 'y', 'ring', 'ring_incriment', 'ring_progress']] = number_data.apply(
    lambda row: pd.Series(get_coordinates(row)), axis=1)
number_data['prime'] = 0
```

```python
%%time
number_data[['prime',
           'determining_prime',
           'lpf',
           'digits',
           'first_digit',
           'last_digit',
           'digits_length',
           'mean_digit',
           'median_digit',
           'mode_digit',
           'mode_count'
          ]] = number_data.apply(
    lambda row: pd.Series(check_prime(row)), axis=1)
```

```python
number_data.head()
```

```python
%%time
number_data[[
    'gpf',
    'prime_factors',
    'factor_count',
    'factor_mean',
    'factor_median',
    'factor_mode',
    'factor_mode_count'
]] = number_data.apply(
    lambda row: pd.Series(get_prime_factors(row, number_data)), axis=1
)
number_data.head(50)
```

```python
prediction_number_data = number_data[[
    'number',
    'x',
    'y',
    'ring',
    'ring_incriment',
    'ring_progress',
    'prime',
    'first_digit',
    'last_digit',
    'digits_length',
    'mean_digit',
    'median_digit',
    'mode_digit',
    'mode_count'
]]
```
