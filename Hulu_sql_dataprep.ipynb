```python
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
show_df = pd.read_csv('HuluRaw.csv', index_col=0)
temp_df = pd.DataFrame()
```

```python
show_df.head()
```

```python
col_names = {}
for name in show_df.columns:
    col_names[name] = name.replace('/', '_')
show_df.rename(columns=col_names, inplace=True)
show_df.head()
```

```python
def desired_columns(df):
    results = pd.DataFrame()
    for name in df.columns:
        is_desired = input('Include ' + name + ' ? y/n').lower()
        if is_desired == 'y' or is_desired == 'yes':
            results[name] = df[name]
    return results
```

```python
final_df = desired_columns(show_df)
```

```python
final_df.head()
```

```python
final_df = final_df.drop_duplicates(subset=['show_id'])
```

```python
def obj_var(name):
    try:
        temp_df['test_datetime'] = pd.to_datetime(show_df[name])
    except ValueError:
        temp_df['test_datetime'] = show_df[name]
    is_datetime = pd.api.types.is_datetime64_any_dtype(temp_df['test_datetime'])
    if is_datetime == True:
        final_type = 'DATE'
    else:
        max_length = int(show_df[name].str.len().max() * 2)
        if max_length <= 1000:
            final_type = 'VARCHAR(' + str(max_length) +')'
        elif max_length > 1000:
            final_type = 'TEXT'
        else:
            final_type = obj_var(name)
    return final_type
```

```python
def get_primary(name):
    set_primary = input('Is ' + name + ' the primary key? y/n').lower()
    if set_primary == 'y' or 'yes':
        return True
    elif set_primary == 'n' or 'no':
        return False
    else:
        return get_primary()
```

```python
def variable_info(name, var_type, primary_set):
    final_type = ''
    set_primary = False
    if var_type == 'int64':
        final_type = 'INTEGER'
    elif var_type == 'float64':
        final_type = 'DOUBLE PRECISION'
    elif var_type == 'bool':
        final_type = 'BOOLEAN'
    elif var_type == 'object':
        final_type = obj_var(name)
    else:
        final_type = var_type.upper()
    if primary_set == False:
        set_primary = get_primary(name)
        if set_primary:
            final_type = final_type + ' PRIMARY KEY'
    final_type = final_type + ','
    return final_type, set_primary
        
                
```

```python
def assemble_instructions(df):
    table_name = input('Name of table?')
    instructions = 'CREATE TABLE ' + table_name + ' ('
    primary_set = False
    for name in df.columns:
        var_type, set_primary = variable_info(name, str(df.dtypes[name]), primary_set)
        new_var = name + " " + var_type
        instructions = instructions + '\n\t' + new_var
        if set_primary == True : primary_set = True
    instructions = instructions + '\n);'
    return instructions
```

```python
show_df = show_df.drop(columns = ['Unnamed: 181'])
```

```python
print(assemble_instructions(final_df))
```
