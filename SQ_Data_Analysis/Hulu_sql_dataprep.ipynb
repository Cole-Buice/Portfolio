```python
# Load raw data
show_df = pd.read_csv('HuluRaw.csv', index_col=0)
```

```python
# Rename column names to fit a naming convention.
col_names = {}
for name in show_df.columns:
    col_names[name] = name.replace('/', '_')
show_df.rename(columns=col_names, inplace=True)
show_df.head()
```

```python
# Create tool for easily selecting which columns are important.
class ColumnSelectorApp:
    def __init__(self, df):
        self.df = df # Initial dataframe being selected from
        self.number_of_columns = 2 # Initial number of columns to display
        self.selected_columns = [] # Store selected columns
        self.new_df = None  # Attribute to store the new DataFrame

        # Create title
        self.title = widgets.HTML('<h2 style="text-align:center;">Select desired variables</h2>') # Title for tool

        # Create slider widget for adjusting columns
        self.slider = widgets.IntSlider(value=self.number_of_columns, min=1, max=10, description='Columns:', continuous_update=False)
        
        # Register the event handler function
        self.slider.observe(self.update_layout, names='value')

        # Create checklist container
        self.checklist_container, self.checkboxes = self.create_checklist(self.df, columns=self.number_of_columns)

        # Create "Finished" button
        self.finish_button = widgets.Button(description="Finished")
        self.finish_button.on_click(lambda button: self.on_finish_button_clicked(self.df, button))

        # Display the widgets
        display(widgets.VBox([self.title, self.slider, self.checklist_container, self.finish_button]))
        
    # Create a check box for each column in the initial dataframe and put them in a container
    def create_checklist(self, df, max_width='95%', columns=3):
        checkboxes = {}
        for column in df.columns:
            checkboxes[column] = widgets.Checkbox(description=column, layout=widgets.Layout(width='auto', flex='auto'))
            container = widgets.GridBox(children=list(checkboxes.values()), layout=widgets.Layout(grid_template_columns=f'repeat({columns}, auto)', width=max_width))
        return container, checkboxes

    # When the sliger adjusts number of columns update the layout to match
    def update_layout(self, change):
        new_columns = change.new
        self.number_of_columns = new_columns
        self.checklist_container.layout.grid_template_columns = f'repeat({new_columns}, auto)'

    # When the finished button is clicked update the "new_df" variable to store the final trimmed dataframe.
    def on_finish_button_clicked(self, df, button):
        self.selected_columns = [col[0] for col in self.checkboxes.items() if col[1].value]
        self.new_df = self.df[self.selected_columns]
```

```python
# Select desired columns
desired_columns = ColumnSelectorApp(show_df)
```

```python
# Create new dataframe from desired columns
final_df = desired_columns.new_df
# View new dataframe
final_df.head()
```

```python
# Remove all duplicate values.
final_df = final_df.drop_duplicates(subset=['show_id'])
```

```python
# Get additional details about object variables.
def obj_var(name, df):
    # Check if variable contains date time objects.
    temp_df = pd.DataFrame()
    try:
        temp_df['test_datetime'] = pd.to_datetime(df[name]) # Attempt to convert the objects to a date time in a sepperate dataframe.
    except ValueError:
        temp_df['test_datetime'] = df[name]
    is_datetime = pd.api.types.is_datetime64_any_dtype(temp_df['test_datetime']) # Check if the objects converted to a date time
    if is_datetime == True:
        final_type = 'DATE' # If date time type is confirmed the final SQL type is DATE
    else:
        max_length = int(df[name].str.len().max() * 2) # Get the largest string length of the object and double it in case future string lengths increase.
        # Based on the max length determine if the string type should be VARCHAR or TEXT
        if max_length <= 1000:
            final_type = 'VARCHAR(' + str(max_length) +')'
        elif max_length > 1000:
            final_type = 'TEXT'
    return final_type
```

```python
# Find out if the variable is the primary key.
def get_primary(name):
    set_primary = input('Is ' + name + ' the primary key? y/n').lower() # Request if the given variable should be set to the primary key.
    # Return boolean value if correctly answered. If not ask again.
    if set_primary == 'y' or set_primary == 'yes':
        return True
    elif set_primary == 'n' or set_primary == 'no':
        return False
    else:
        return get_primary()
```

```python
# Get additional info about the variables to get accurate data types and primary key.
def variable_info(name, var_type, primary_set, not_final, df):
    final_type = '' # final code for variable
    set_primary = False # Is this the primary key? Has the primary key already been set?
    type_dict = {
        'int64': 'INTEGER',
        'float64': 'DOUBLE PRECISION',
        'bool': 'BOOLEAN',
    } # Simple type dictionary
    # get SQL type
    if var_type in type_dict:
        final_type = type_dict[var_type]
    elif var_type == 'object':
        final_type = obj_var(name, df) # Object is a special case run extra check.
    else:
        final_type = var_type.upper() # If type not found set data type to var_type all upper case. Add types to function as they present themselves.
    if primary_set == False:
        set_primary = get_primary(name) # Check if this is the primary key if one has not yet been set.
        if set_primary:
            final_type = final_type + ' PRIMARY KEY' # If needed adjust code to a primary key.
    final_type = final_type + ',' if not_final else final_type # Finally finish of code with a comma if needed.
    return final_type, set_primary
        
                
```

```python
# Initiate creation of SQL data table creation code for the dataframe.
def assemble_instructions(df):
    table_name = input('Name of table?') # Request a name for the data table
    instructions = 'CREATE TABLE ' + table_name + ' (' # Initial Code
    primary_set = False # Has a primary key been selected.
    for index, name in enumerate(df.columns):
        final = index == len(df.columns) - 1
        var_type, set_primary = variable_info(name, str(df.dtypes[name]), primary_set, not final, df) # Get variable type and primary key info. Also pass if the variable is the final in the dataframe for propper commas.
        new_var = name + " " + var_type # Generate code for the variable
        instructions = instructions + '\n\t' + new_var # Update code instructions with a new line and tab.
        if set_primary == True : primary_set = True # Update primary_set if changed
    instructions = instructions + '\n);' # Close the table creation code.
    return instructions
```
