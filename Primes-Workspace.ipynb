```python
prime_data = pd.DataFrame()
prime_data['number'] = range(2, 500000)
prime_data.head()
```

```python
%%time
prime_data[['x', 'y']] = prime_data.apply(lambda row: pd.Series(get_coordinates(row)), axis=1)
one_data = pd.DataFrame({'number': [1], 'x': [0], 'y': [0]})
prime_data = pd.concat([one_data, prime_data], ignore_index=True)
prime_data['prime'] = 0
```

```python
%%time
prime_data[['prime', 'lpf', 'digits', 'last_digit', 'digits_length']] = prime_data.apply(lambda row: pd.Series(check_prime(row)), axis=1)
```

```python
%%time
plt.figure(figsize=(10, 10))
sns.scatterplot(data=prime_data[prime_data['prime'] == 1],
                x='x',
                y='y',
                hue='prime',
                s=2
               )

plt.show()
```

```python
prime_data.head(50)
```

```python
%%time
prime_data[['gpf', 'prime_factors']] = prime_data.apply(lambda row: pd.Series(get_prime_factors(row)), axis=1)
```

```python
%%time
prime_data['factor_count'] = prime_data['prime_factors'].apply(len)
```

```python
prime_data.head(50)
```

```python
factor_counts = prime_data.groupby('factor_count').size().reset_index(name='count')
```

```python
sns.scatterplot(data=factor_counts,
            x='factor_count',
            y='count'
           )
plt.show()
```

```python
plt.figure(figsize=(15, 15))
scatter = sns.scatterplot(data=prime_data.iloc[:100000],
                x='x',
                y='y',
                hue='factor_count',
                #palette='bright',
                s=8,
                legend=False,
                marker='s',
               )

scatter.grid(False)

plt.show()
```

```python
def plot_data(rows):
    plt.figure(figsize=(15, 15))
    scatter = sns.scatterplot(data=prime_data.iloc[:rows],
                x='x',
                y='y',
                hue='factor_count',
                #palette='bright',
                s=2,
                legend=False,
                marker='s',
               )
    scatter.grid(False)
    plt.show()

rows_selector = widgets.IntSlider(min=1, max=len(prime_data), value=len(prime_data), description='Rows:')
widgets.interactive(plot_data, rows=rows_selector)
```
