```python
# Import Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
import sklearn.metrics as metrics
from sklearn.metrics import make_scorer, recall_score, f1_score, precision_score
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from xgboost import plot_importance
from sklearn.model_selection import train_test_split, PredefinedSplit, GridSearchCV
```

```python
# Load in data as pandas dataframe
prime_prediction_df = pd.read_csv('prediction_number_data.csv')
```

```python
# Check Data
prime_prediction_df.head()
```

```python
# Data cleaning is not needed as this data was constructed and verified first hand
# EDA has been done previously in the Primes-Workspace file. The data was itself also constructed via EDA of the qualities of numbers.
```

# Prepare data for machine learning models
Primes are far more rare than composit numbers and are therefore undersampled in the dataset. To rectrify this I will be pulling a sample from the full data set that balances the ratio of primes and composits.

```python
# Create a balanced dataset
# Split data into prime and composit dataframes.
prime_set = prime_prediction_df[prime_prediction_df['prime'] == 1]
composit_set = prime_prediction_df[prime_prediction_df['prime'] == 0]
# Get a random sample of composit numbers the same size as the prime sample total.
composit_sample = composit_set.sample(n=prime_set['number'].count(), random_state=123)
# Combine the prime dataframe with the composit sample dataframe and display new balanced dataframe.
balanced_set = pd.concat([prime_set, composit_sample], ignore_index=True)
balanced_set.describe()
```

```python
#Construct train test split data

X = balanced_set.drop(['prime'], axis=1)

y = balanced_set['prime']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=13, stratify=y)
```

```python
# Fit a logistic regression to the data
clf = LogisticRegression()
clf.fit(X_train,y_train)
```

```python
# Model coefficient values
clf.coef_
```

```python
# Model intercept
clf.intercept_
```

```python
# Visualize coefficients

coefficients = clf.coef_[0]
feature_names = X.columns

coef_df = pd.DataFrame({'Coefficient': coefficients, 'Feature': feature_names})

coef_df = coef_df.sort_values(by='Coefficient', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Coefficient', y='Feature', data=coef_df)
plt.xlabel('Coefficient')
plt.ylabel('Feature')
plt.title('Logistic Regression Coefficients')
plt.show()
```

```python
coef_df
```

```python
# Test predictions and gather metric data

y_pred = clf.predict(X_test)

print("Accuracy:", "%.6f" % metrics.accuracy_score(y_test, y_pred))
print("Precision:", "%.6f" % metrics.precision_score(y_test, y_pred))
print("Recall:", "%.6f" % metrics.recall_score(y_test, y_pred))
print("F1 Score:", "%.6f" % metrics.f1_score(y_test, y_pred))
```

```python
# Visualize predictive power using a confusion matrix
cm = metrics.confusion_matrix(y_test, y_pred, labels = clf.classes_)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix = cm,display_labels = clf.classes_)
disp.plot()
```

With the given variables a logistic regression model is not reliable for predicting primes.
It mannaged to get zero false positives but also had zero true positives.
Essentially the model assumed that no numbers are prime.

```python
# Create a function for storing model scores for comparison

def get_model_scores(model_type, test, pred):
    # Get model scores
    ac_score = metrics.accuracy_score(test, pred)
    pc_score = metrics.precision_score(test, pred)
    rc_score = metrics.recall_score(test, pred)
    f1_score = metrics.f1_score(test, pred)
    # Create and return a dataframe with the models type and its scores as values.
    score_df = pd.DataFrame(
        {'Model': [model_type],
        'F1':  [f1_score],
        'Recall': [rc_score],
        'Precision': [pc_score],
        'Accuracy': [ac_score]
        }
    )
    return score_df
```

```python
# Store logistic regression model scores
# Create scoreboard dataframe
model_scores = pd.DataFrame()
# Add model scores to the scoreboard.
model_scores = pd.concat(
    [model_scores, get_model_scores('LogisticRegression', y_test, y_pred)],
    ignore_index=True)
# View model scores
model_scores.head()
```

```python
# Random forest
# New train test split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, stratify=y, random_state = 4444)
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size = 0.20, stratify=y_train, random_state = 60)
```

```python
# Create random forest
rf = RandomForestClassifier(random_state=0)

# Determine cross validation parameters
cv_para = {
    'n_estimators': [50, 100],
    'max_depth': [5, None],
    'min_samples_split': [5, 10, 15],
    'min_samples_leaf': [1, 5],
    'max_features': ['sqrt', 'log2', None],
    'bootstrap': [True]
}

# Determine scoring parameters
scoring = {'accuracy', 'precision', 'recall', 'f1'}

# Create grid search cross validation to find optimal parameters.
rf_val = GridSearchCV(rf, cv_para, refit='f1', n_jobs = -1, verbose = 1)
```

```python
%%time
# Train the grid search
rf_val.fit(X_train, y_train)
```

```python
# Get best parameters from the grid search
rf_val.best_params_
```

```python
# Create new random forest using the best parameters.
rf_opt = RandomForestClassifier(n_estimators = 50, max_depth = 5, 
                                min_samples_leaf = 1, min_samples_split = 5,
                                max_features = 'sqrt', bootstrap = True, random_state = 0)
```

```python
# Train final model.
rf_opt.fit(X_train, y_train)
```

```python
# Test model
y_pred = rf_opt.predict(X_test)
```

```python
# Visualize predictive power using a confusion matrix.
cm = metrics.confusion_matrix(y_test, y_pred, labels = clf.classes_)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix = cm,display_labels = clf.classes_)
disp.plot()
```

```python
# Add model score to scoreboard.
model_scores = pd.concat(
    [model_scores, get_model_scores('RandomForest', y_test, y_pred)],
    ignore_index=True)
# Display scoreboard.
model_scores.head()
```

Unlike the logistic regression model the random forest model was actually useful at predicting primes using the given variables. Each metric scored above 0.7 which considering the randomness of primes is a great score.

```python
# XGBoost Model

# Create new train test split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5, stratify=y, random_state = 25)
```

```python
# Create XGBoost clasifier.
xgb = XGBClassifier(objective='binary:logistic', random_state=0)
```

```python
# Determine cross validation parameters.
cv_params = {
    'n_estimators': [10, 30, 50],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 4],
    'min_child_weight': [1, 5],
    'subsample': [0.7, 0.8],
    'colsample_bytree': [0.7, 0.8],
    'gamma': [0, 0.1],
    'lambda': [0, 0.1],
    'alpha': [0, 0.1],
    'objective': ['binary:logistic'],
    'eval_metric': ['logloss'],
}
```

```python
# Determine scoring paramaters
scoring = {
    'recall': make_scorer(recall_score, zero_division=1),
    'f1': make_scorer(f1_score, zero_division=1),
    'accuracy': 'accuracy',
    'precision': make_scorer(precision_score, zero_division=1)
}

# Create a cross validation grid search.
xgb_cv = GridSearchCV(xgb,
                      cv_params,
                      scoring = scoring,
                      cv = 3,
                      refit = 'f1',
                      verbose = 10
                     )
```

```python
%%time
# Train the cross validation.
xgb_cv = xgb_cv.fit(X_train, y_train)
```

```python
# Get best paramaters from the cross validation
xgb_cv.best_params_
```

```python
# Create new XGBoost classifier using the optimized paramaters.
xgb_opt = XGBClassifier(
    alpha= 0.1,
    colsample_bytree= 0.7,
    eval_metric= 'logloss',
    gamma= 0,
    reg_lambda= 0,
    learning_rate= 0.1,
    max_depth= 4,
    min_child_weight= 1,
    n_estimators= 50,
    objective='binary:logistic',
    subsample= 0.8)
```

```python
# Train the optimized model.
xgb_opt.fit(X_train, y_train)
```

```python
# Test the optimized model
y_pred = rf_opt.predict(X_test)
```

```python
# Visualize the predictive power of the model using a confusion matrix.
cm = metrics.confusion_matrix(y_test, y_pred, labels = clf.classes_)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix = cm,display_labels = clf.classes_)
disp.plot()
```

```python
# Add model to scoreboard
model_scores = pd.concat(
    [model_scores, get_model_scores('XGBoost', y_test, y_pred)],
    ignore_index=True)
# Display scoreboard
model_scores.head()
```

The XGBoost model was also successful and was the most successful model so far with a slight improvement from the random forest over every observed metric.

```python
# Visualize variable importance.
plot_importance(xgb_opt)
```

When we observe the feature importances of the model we can see that the last digit is the most important variable. This isnt too surprising since all primes except 2 and 5 must end in 1, 3, 7, or 9. This allows the model to reduce potential primes by 60%. The second most important variable was the mean digit. This was actually surprising. I'm personally unaware of a particular reason that the average digit would suggest a number is prime especially so strongly. This could be an interesting insight to dive into further in a future EDA and could possibly be leveraged to build a more powerful model. The mode digit is third most important and this is also surprising however its level of importance is considerably less than the mean digit, so while its interesting its priority of further research is not as significant as the mean digit. Number and ring progress follow closely behind mode digit maintaing an F Score above 30 after that scores drop to the low 20s and below.
