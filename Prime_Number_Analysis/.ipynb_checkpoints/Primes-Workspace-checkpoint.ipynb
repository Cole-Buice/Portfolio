```python
# Create a new data frame
number_data = pd.DataFrame()
# Populate the data frame with a 'number' column containing a range of numbers.
number_data['number'] = range(1, 500001)
```

```python
%%time
# Extract Ulam coordinate data from the numbers.
number_data[['x', 'y', 'ring', 'ring_incriment', 'ring_progress']] = number_data.apply(
    lambda row: pd.Series(get_coordinates(row)), axis=1)
# Create a 'prime column and set them equal to zero'
number_data['prime'] = 0
```

```python
%%time
# Extract prime and digit data.
number_data[['prime',
           'determining_prime',
           'lpf',
           'digits',
           'first_digit',
           'last_digit',
           'digits_length',
           'mean_digit',
           'median_digit',
           'mode_digit',
           'mode_count'
          ]] = number_data.apply(
    lambda row: pd.Series(check_prime(row)), axis=1)
```

```python
class ScatterPlotter:
    def __init__(self, data):
        self.data = data
        self.figsize = (20, 20)
        self.marker_size = 3
        self.marker_style = 's'
        self.x_selector = widgets.Dropdown(options=self.data.columns, description='X:')
        self.y_selector = widgets.Dropdown(options=self.data.columns, description='Y:')
        self.hue_selector = widgets.Dropdown(options=self.data.columns, description='Hue:')
        self.legend = widgets.Dropdown(options=[True, False], description='Legend:')
        self.marker_size_slider = widgets.FloatSlider(value=self.marker_size, min=0.01, max=100.0, step=0.01, description='Marker Size:')
        self.additional_columns, self.checkboxes = self.create_checkbox_dropdown(df=self.data, description='Special Case Columns:')
        self.inputs_container = widgets.VBox(layout=widgets.Layout(overflow_y='auto'))
        self.plot_button = widgets.Button(description="Plot")
        self.clear_button = widgets.Button(description="Clear Plots")
        self.plot_output = widgets.Output()
        self.plot_button.on_click(self.plot_handler)
        self.clear_button.on_click(self.clear_plots)
        self.observe_checkboxes()
        self.display_widgets()

    def create_checkbox_dropdown(self, df, description):
        checkboxes = {}
        for column in df.columns:
            checkboxes[column] = widgets.Checkbox(description=column, layout=widgets.Layout(width='auto', flex='auto'))
        container = widgets.VBox([
            widgets.Label(description, layout=widgets.Layout(margin='0 0 5px 0')),  # Description label
            widgets.VBox(children=list(checkboxes.values()), layout=widgets.Layout(height='200px', overflow_y='scroll'))
        ])
        return container, checkboxes

    def observe_checkboxes(self):
        for checkbox in self.checkboxes.values():
            checkbox.observe(self.update_input_fields, 'value')
    
    def update_input_fields(self, change):
        self.inputs_container.children = []
        checked_columns = [checkbox.description for checkbox in self.checkboxes.values() if checkbox.value]
        new_widgets = []
        for column in checked_columns:
            variable_type = self.get_variable_type(column)
            input_widget = self.create_input_widget(variable_type)
            input_widget.observe(self.handle_input_change, 'value')
            new_widgets.append(widgets.HBox(children=[widgets.Label(f"{column}:"), input_widget], layout=widgets.Layout(justify_content='space-between')))

        new_widgets = widgets.VBox(new_widgets, layout=widgets.Layout(width='20%', flex='auto'))
        self.inputs_container.children = [new_widgets]

    def get_variable_type(self, column):
        return pd.api.types.infer_dtype(self.data[column])
    
    def create_input_widget(self, variable_type):
        if variable_type == 'string':
            return widgets.Text()
        elif variable_type == 'floating' or variable_type == 'integer':
            return widgets.FloatText() if variable_type == 'floating' else widgets.IntText()
        else:
            return widgets.Text()  # Default to text input for other variable types
    
    def handle_input_change(self, change):
        print("Input changed:", change.new)

    def display_widgets(self):
        self.widgets_display = widgets.VBox([
            self.x_selector,
            self.y_selector,
            self.hue_selector,
            self.legend,
            self.marker_size_slider,
            self.additional_columns,
            self.inputs_container,
            self.plot_button,
            self.clear_button,
            self.plot_output
        ])
        display(self.widgets_display)
        
    def plot_handler(self, button):
        with self.plot_output:
            self.plot(self.x_selector.value, self.y_selector.value, self.hue_selector.value, self.marker_size_slider.value, self.additional_columns.value)
        
    def plot(self, x_column, y_column, hue_column, marker_size, additional_columns, palette=None):
        plt.figure(figsize=self.figsize)
        scatter = sns.scatterplot(data=self.data,
                                  x=x_column,
                                  y=y_column,
                                  hue=hue_column,
                                  palette=palette,
                                  s=marker_size,
                                  legend=self.legend.value,
                                  marker=self.marker_style)
        scatter.grid(False)
        plt.show()
        
    def clear_plots(self, button):
        self.plot_output.clear_output()
```

```python
scatter_plotter = ScatterPlotter(number_data)
```
